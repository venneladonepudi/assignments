```python
import matplotlib.pyplot as plt
subjects=['Python','Java','C++','SQL']
students=[45,38,30,50]

plt.bar(subjects,students)
plt.title("Students Enrolled")
plt.xlabel("Subjects")
plt.ylabel("Number of students")
plt.show()
```

```python
plt.bar(subjects,students,color='red')
plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May']
sales=[25,30,40,35,50]

plt.plot(months,sales)

plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")

plt.show()
```

```python
plt.plot(months,sales,
         marker='o',
         linestyle='--',
         color='blue')
plt.show()
```

```python
study_hours=[1,2,3,4,5,6,7]
marks=[35,40,50,60,68,80,90]

plt.scatter(study_hours,marks)

plt.title("Study hour vs Marks")
plt.xlabel("Study Hours")
plt.ylabel("Marks")

plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.scatter(x,y,marker='D')
plt.title("Diamond marker")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.scatter(x,y,marker='*',s=20)
plt.title("Star marker")
plt.show()
```

```python
labels=['Python','Java','C++','SQL']
students=[40,30,20,10]
plt.pie(students,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90)
plt.title("Students Learning Programming Language")
plt.show()
```

```python
#Create a bar chart showing the number of employees in four departments.
import matplotlib.pyplot as plt


departments=['HR','IT','Finance','Sales']
employees=[2,3,2,4]

plt.bar(departments,employees)
plt.title("Number of employees in four departments")
plt.xlabel("Departments")
plt.ylabel("Employees")
plt.show()
```

```python
#Draw a line chart showing monthly temperatures for six months.
months=['Jan','Feb','Mar','Apr','May','Jun']
temperature=[22,24,28,32,36,30]

plt.plot(months,temperature,marker='o')
plt.title("Monthly temperature for six months")
plt.xlabel("Months")
plt.ylabel("Temperature")
plt.show()
```

```python
#Create a scatter plot showing age versus salary.
age=[23,25,30,35,40]
salary=[30000,40000,50000,60000,70000]
plt.scatter(age,salary)
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
#Draw a bar chart showing sales of five products
products=['Laptop','Mobile','Tablet','Headphones','Smartwatch']
sales=[85,150,60,95,110]
plt.bar(products,sales)
plt.title("Sales of five products")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()
```

```python
#Plot a line chart showing the marks obtained in five students
tests = ['Test 1', 'Test 2', 'Test 3', 'Test 4', 'Test 5']
marks = [72, 80, 76, 89, 84]
plt.plot(tests, marks, marker='o')
plt.title("Marks Obtained in Five Tests")
plt.xlabel("Tests")
plt.ylabel("Marks")
plt.show()
```
