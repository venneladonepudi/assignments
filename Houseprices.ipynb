level2:


```python
#find all columns containing missing values
print(df.columns[df.isnull().any()])
```

```python
#calculate the percentage of missing values of each column
missing_percent=(df.isnull().sum()/len(df))*100
print(missing_percent)
```

```python
#identify the column with the highest number of missing values
print(df.isnull().sum().idxmax())
```

```python
#fill missing values in grvlivarea usimg the median
df["GrLivArea"]=df["GrLivArea"].fillna(df["GrLivArea"].median())
```

```python
#fill missing values in neghbourhood using the mode
df["Neighborhood"]=df["Neighborhood"].fillna(df["Neighborhood"].mode()[0])
```

```python
#drop columns where more than 50% values are missing
threshold=len(df)*0.5
df=df.dropna(axis=1,thresh=threshold)
print(df.isnull().sum())
```
